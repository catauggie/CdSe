```python
import matplotlib.pyplot as plt

results_df = pd.read_csv('parameters_over_generations_Varshni_B1.csv')
# Assume results_df is the DataFrame with columns: 'E_0', 'a1', 'a2', 'R^2'

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))

# Plot E_0
axes[0].plot(results_df[0:24].index, results_df['E_0'][0:24], marker='o', linestyle='-')
axes[0].set_title('$E_0$ over Generations')
axes[0].set_xlabel('Generation')
axes[0].set_ylabel('$E_0$')
axes[0].grid()

# Plot a1
axes[1].plot(results_df[0:24].index, results_df['a1'][0:24], marker='o', linestyle='-')
axes[1].set_title('$a_1$ over Generations')
axes[1].set_xlabel('Generation')
axes[1].set_ylabel('$a_1$')
axes[1].grid()

# Plot a2
axes[2].plot(results_df[0:40].index, results_df['a2'][0:40], marker='o', linestyle='-')
axes[2].set_title('$a_2$ over Generations')
axes[2].set_xlabel('Generation')
axes[2].set_ylabel('$a_2$')
axes[2].grid()

# Plot R^2
axes[3].plot(results_df[8:20].index, results_df['R^2'][8:20], marker='o', linestyle='-')
axes[3].set_title('$R^2$ over Generations')
axes[3].set_xlabel('Generation')
axes[3].set_ylabel('$R^2$')
axes[3].grid()

# Adjust layout
plt.tight_layout()

# Save the figure (optional)
plt.savefig('parameters_over_generations_Varshni_B1.pdf')

# Show the figure
plt.show()
```

```python
history_df = pd.read_csv('parameters_history.csv')


# Plotting the history
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))

# Plot E_0
axes[0].plot(history_df[0:24].index, history_df['E_0'][0:24], marker='o', linestyle='-')
axes[0].set_title('$E_0$ over Generations')
axes[0].set_xlabel('Generation')
axes[0].set_ylabel('$E_0$')
axes[0].grid()

# Plot a
axes[1].plot(history_df[0:30].index, history_df['a'][0:30], marker='o', linestyle='-')
axes[1].set_title('a over Generations')
axes[1].set_xlabel('Generation')
axes[1].set_ylabel('a')
axes[1].grid()

# Plot theta
axes[2].plot(history_df.index, history_df['theta'], marker='o', linestyle='-')
axes[2].set_title(r'$\theta$ over Generations')
axes[2].set_xlabel('Generation')
axes[2].set_ylabel(r'$\theta$')
axes[2].grid()

# Plot R^2
axes[3].plot(history_df[4:26].index, history_df['R^2'][4:26], marker='o', linestyle='-')
axes[3].set_title('$R^2$ over Generations')
axes[3].set_xlabel('Generation')
axes[3].set_ylabel('$R^2$')
axes[3].grid()

# Adjust layout
plt.tight_layout()

# Save the figure (optional)
plt.savefig('parameters_over_generations_Fan_B1.pdf')

# Show the figure
plt.show()
```
